## Title : Image to Pencil Sketch with Python

```python
# Importing the required Libraries

import matplotlib.pyplot as plt
import cv2
```

```python
# Reading the Original Image

img = cv2.imread('picture.jpg')
org_img =cv2.resize(img, (700, 500))

plt.imshow(org_img)
plt.show()
```

```python
# Converting the RGB Image to Gray-Scale Image

grsc_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)

plt.imshow(grsc_img)
```

```python
# Inverting the Gray-Scale Image

inv_grsc_img = cv2.bitwise_not(grsc_img)

plt.imshow(inv_grsc_img)
```

```python
# Applying a Gaussian Blur to the Inverted Gray-Scale Image

invblur_img = cv2.GaussianBlur(inv_grsc_img, (21, 21),sigmaX=0, sigmaY=0)

plt.imshow(invblur_img)
```

```python
# Mixing the Gray-Scale Image with Inverted Blurry Image

def gray_InvBlur(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch_img = gray_InvBlur(grsc_img, invblur_img)

plt.imshow(pencil_sketch_img)
```

```python
# ORIGINAL IMAGE AND PENCIL SKETCH IMAGE

plt.imshow(org_img)

plt.imshow(pencil_sketch_img)
```

```python
# Saving and displaying the Final Pencil Sketch Image

cv2.imwrite('Pencil_Sketch.jpg',pencil_sketch_img)
```
